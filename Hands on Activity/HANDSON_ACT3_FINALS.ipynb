**Import**

```python
import numpy as np
import matplotlib.pyplot as plt
```

**List of maps**

```python
maps = [
    "Blue Zenith",
    "Freedom Dive",
    "Camellia",
    "Sidetracked Day",
    "The Big Black",
    "Galaxy Collapse",
    "Brain Power"
]
```

**List of Play Counts**

```python
playcounts = [1450, 2100, 1750, 900, 2300, 1600, 1950]
```

**Dictionaries of difficulty**

```python
difficulty = {
    "Blue Zenith": 5.5,
    "Freedom Dive": 6.2,
    "Camellia": 6.8,
    "Sidetracked Day": 4.9,
    "The Big Black": 7.0,
    "Galaxy Collapse": 7.5,
    "Brain Power": 6.6
}

diff_names = list(difficulty.keys())
diff_values = list(difficulty.values())
```

**Dictionaries of accuracy**

```python
accuracy = {
    "Blue Zenith": 97.5,
    "Freedom Dive": 95.8,
    "Camellia": 96.2,
    "Sidetracked Day": 98.0,
    "The Big Black": 93.4,
    "Galaxy Collapse": 94.6,
    "Brain Power": 96.8
}

acc_names = list(accuracy.keys())
acc_values = list(accuracy.values())
```

```python
maps, playcounts, difficulty, accuracy
```

### **Matrices**

```python
ma = np.array([950, 870, 1000, 780, 1100, 900, 1030]) #Scores
mb = np.array([975, 982, 968, 960, 990, 985, 978]) #Accuracy
mc = np.array([850, 920, 910, 880, 1000, 970, 940]) #High Score on Big Black
md = np.array(diff_values) #Difficulty Values
me = np.array(acc_values) #Accuracy Values
```

### **Visualization and Questions**

**1. How many times are the popular OSU beatmaps played based on the recorded playcounts?**

```python
plt.figure(figsize=(5, 3))
plt.bar(maps, playcounts)
plt.title("OSU Play Counts per Beatmap")
plt.xlabel("Beatmap")
plt.ylabel("Play Count")
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

**2. What are the player scores recorded from 7 attempts on the beatmap Blue Zenith?**

```python
plt.figure(figsize=(5, 3))
plt.plot(ma, marker="8")
plt.title("Player Score on Blue Zenith")
plt.xlabel("Attempt")
plt.ylabel("Score (x1000)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

**3. What are the accuracy values (x100) recorded from 7 attempts on the beatmap Blue Zenith?**

```python
plt.figure(figsize=(5, 3))
plt.plot(mb, marker="8")
plt.title("Accuracy on Blue Zenith (x100)")
plt.xlabel("Attempt")
plt.ylabel("Accuracy (x100)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

**4. What are the high scores achieved by the player on the beatmap The Big Black?**

```python
plt.figure(figsize=(5, 3))
plt.scatter(range(len(mc)), mc, marker="x")
plt.title("High Scores on The Big Black")
plt.xlabel("Play Index")
plt.ylabel("Score (x1000)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

**5. What are the average difficulty star ratings of the selected OSU beatmaps?**

```python
plt.figure(figsize=(5, 3))
plt.bar(diff_names, md)
plt.title("OSU Beatmap Difficulty Rating (Star Level)")
plt.xlabel("Beatmap Name")
plt.ylabel("Difficulty (Stars)")
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

**6. What are the average accuracy percentages of the selected OSU beatmaps?**

```python
plt.figure(figsize=(5, 3))
plt.plot(acc_names, me, marker="8", linestyle="solid")
plt.title("Average Accuracy per OSU Beatmap")
plt.xlabel("Beatmap Name")
plt.ylabel("Accuracy (%)")
plt.ylim(90, 100)
plt.grid(True)
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```
